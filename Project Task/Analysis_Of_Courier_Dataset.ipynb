## Find out total weights in the Orders Dataset

```python
Orders['Total_Weight'] = Orders['SKU'].apply(lambda x: Sku[Sku['SKU'] == x]['Weight (g)'].sum())
```

```python
# SKU se weight mapping create karna
sku_weight = Sku.groupby('SKU')['Weight (g)'].sum().to_dict()
```

```python
sku_weight
```

```python
# Orders DataFrame mein SKU ke according weight map karna
Orders['Total_Weight'] = Orders['SKU'].apply(lambda x: sku_weight.get(x, 0))
```

## Analyze Charges Billed by Courier Company 

```python
Delivery_Zone = Invoices.groupby("Zone")['Billing Amount (Rs.)'].sum().to_dict()
```

```python
Invoices['Charges Billed by Courier Company (Rs.)'] = Invoices['Zone'].apply(lambda x: Delivery_Zone.get(x, 0))
```

```python
Invoices
```

## Find out Zones charged weight

```python
find_zone = Invoices.groupby('Zone')['Charged Weight'].sum().to_dict()
```

```python
Invoices['Zone_Category_Wiz_Total_Weight'] = Invoices['Zone'].apply(lambda x: find_zone.get(x, 0))
```

```python
Invoices
```

```python
Invoices
```

## Find out total amount of zones

```python
find_zone_amount = Invoices.groupby('Zone')['Billing Amount (Rs.)'].sum().to_dict()
```

```python
Invoices['Total_Amount_Zones'] = Invoices['Zone'].apply(lambda x: find_zone_amount.get(x, 0))
```

## Find out Expected rates

```python
Expected_Rates = Couries_Rates.groupby([Invoices['Zone']])['Weight Slabs'].sum().to_dict()
```

```python
Expected_Rates
```

```python
Invoices['Expected Rates'] = Invoices['Zone'].apply(lambda x: Expected_Rates.get(x, 0))
```

```python
Invoices.head()
```

## Find out Total amount of courier

```python
Invoices['Total_Amount_By_Courier'] = Invoices['Total_Amount_Zones'] - Invoices['Expected Rates']
```
